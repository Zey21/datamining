# KNN, Bayes, Decision Tree

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from collections import OrderedDict
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, precision_score, f1_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
```

```python
url_dataset = "https://raw.githubusercontent.com/Zey21/dataset/main/BankChurners.csv"
dataset = pd.read_csv(url_dataset)
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
list_drop = ['CLIENTNUM','Attrition_Flag','Card_Category','Total_Relationship_Count','Months_Inactive_12_mon','Contacts_Count_12_mon','Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1','Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2']
dataset = dataset.drop(columns = list_drop)
```

Mengolah data

```python
data1 = dataset[["Gender","Education_Level"]]
```

```python
data1.head()
```

```python
data2 = dataset[["Marital_Status","Income_Category"]]
```

```python
data2.head()
```

Mengubah isi text to numeric menggunakan dummies dari pandas

```python
data1 = pd.get_dummies(data1)
data1
```

```python
data2 = pd.get_dummies(data2)
data2
```

Normalisasi Data

```python
dataNormalisasi = dataset.drop(columns = ["Gender","Education_Level","Marital_Status","Income_Category"])
```

```python
dataNormalisasi
```

Normalisasi menggunakan MinMaxScaler

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(dataNormalisasi)
```

```python
HasilNormalisasi = scaler.transform(dataNormalisasi)
```

Mentranformasi Hasil Normalisasi dengan menggunakan Columns Data Normalisasi

```python
HasilNormalisasi = pd.DataFrame(HasilNormalisasi)
HasilNormalisasi.columns = ["Customer_Age","Dependent_count","Months_on_book","Credit_Limit","Total_Revolving_Bal","Avg_Open_To_Buy","Total_Amt_Chng_Q4_Q1","Total_Trans_Amt","Total_Trans_Ct","Total_Ct_Chng_Q4_Q1","Avg_Utilization_Ratio"]
```

```python
HasilNormalisasi
```

Merger 4 tabels with concat from pandas 

```python
TakeCategory = pd.read_csv(url_dataset)
TakeCategory = TakeCategory[["Card_Category"]]
```

```python
Hasil = pd.concat([TakeCategory,data1,data2,HasilNormalisasi],axis =1)
```

```python
Hasil
```

Mengambil data 30%

```python
percent_amount_of_test_data = 0.3
```

Menghitung Data

```python
mat_x = Hasil.iloc[:,1:31].values
mat_y = Hasil.iloc[:,0].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(mat_x, mat_y, test_size = percent_amount_of_test_data, random_state=0)
```

Membuat DataFrame untuk mat_x dan mat_y

```python
DF_mat_x = pd.DataFrame(mat_x)
DF_mat_y = pd.DataFrame(mat_y)
```

```python
DF_mat_x
```

```python
DF_mat_y
```

Memanggil modul untuk menghitung akurasi

```python
model_akurasi = OrderedDict()
```

KNN

```python
K = 15
KNN = KNeighborsClassifier(n_neighbors = K)

Model_KNN = KNN.fit(X_train,y_train)
```

```python
Y_prediksi_knn = KNN.predict(X_test)
DF_prediksi_knn = pd.DataFrame(Y_prediksi_knn)
DF_prediksi_knn
```

```python
HasilKNN = (KNN.score(X_test, y_test) ) * 100
print("Hasil akurasi dari KNN sebesar {} %".format(HasilKNN))
```

Naive Bayes

```python
NBC = GaussianNB()
Model_NBC = NBC.fit(X_train, y_train)
Y_pred_nb = NBC.predict(X_test)
DF_prediksi_bayes = pd.DataFrame(Y_pred_nb)
DF_prediksi_bayes
```

```python
HasilNBC = (NBC.score(X_test, y_test)) * 100
print("Hasil akurasi dari Naive Bayes Classifier sebesar {} %".format(HasilNBC))
```

Decision Tree

```python
DTC = DecisionTreeClassifier(criterion = 'entropy', random_state = 28)
Model_DTC = DTC.fit(X_train, y_train)
DTC
Y_pred_dc = DTC.predict(X_test)
```

```python
DF_prediksi_dtree = pd.DataFrame(Y_pred_dc)
DF_prediksi_dtree
```

```python
HasilDTC = (DTC.score(X_test, y_test)) * 100
print("Hasil akurasi dari Decision Tree Classifier sebesar {} %".format(HasilDTC))
```
