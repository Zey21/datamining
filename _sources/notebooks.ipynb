### Menghitung Dissimilarity

```python
def diss (a,b):
  n=0
  m=2
  for col in ['OvertimePay','OtherPay']:
    if data[col][a]== data[col][b]:
      n+=1
  return (m-n)/m
```

```python
print(diss(1,1))
print(diss(2,1))
```
