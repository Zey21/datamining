```python
dataset = "https://raw.githubusercontent.com/Zey21/dataset/main/Iris.csv"
data = pd.read_csv(dataset)
```

```python
data.head()
```

```python
from sklearn.preprocessing import KBinsDiscretizer
```

```python
sepalLength = data.SepalLengthCm
sepalWidth = data.SepalWidthCm
petalLength = data.PetalLengthCm
petalWidth = data.PetalWidthCm
spesies = data.Species
```

```python
aa = sepalLength.values.reshape(-1,1)
ab = sepalWidth.values.reshape(-1,1)
ac = petalLength.values.reshape(-1,1)
ad = petalWidth.values.reshape(-1,1)
sp = spesies.values.reshape(-1,1)
```

```python
est = KBinsDiscretizer(n_bins=4, encode='ordinal', strategy='uniform')
```

```python
est.fit(aa)
est.fit(ab)
est.fit(ac)
est.fit(ad)
```

```python
xa = est.transform(aa)
xb = est.transform(ab)
xc = est.transform(ac)
xd = est.transform(ad)
```

```python
dataa1 = {'sepal_length':xa}
dataa2 = {'sepal_length':xb}
dataa3 = {'sepal_length':xc}
dataa4 = {'sepal_length':xd}
```
